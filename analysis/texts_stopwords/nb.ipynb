```python
import pandas as pd
from nltk import word_tokenize
```

```python
df = pd.read_csv('../../test_data.csv')
```

```python
texts=df.Output.copy()
texts = texts.apply(lambda x: [word.lower() for word in word_tokenize(x) if word not in [".", ",", "?", '!', ':']])
```

```python
words_from_texts = []
for text in texts.values:
    for word in text:
        words_from_texts.append(word)
```

```python
from collections import Counter

counter = Counter(words_from_texts)
most_common = [x for x,y in counter.most_common(50)]
most_common
```
